```python
from duckduckgo_search import DDGS


results = DDGS().text(
    keywords="python",
    region="wt-wt",
    safesearch="Moderate",
    max_results=10,
    timelimit=3
)

print(results)
```

```python
import requests

# Your local instance URL
SEARX_URL = "http://localhost:8080"

# Example search query
params = {
    "q": "Singapore",
    "format": "json",
    "language": "en",
    "locale": "en_US",
    "categories": "general",
    "pageno": 1
}

response = requests.get(f"{SEARX_URL}/search", params=params)

if response.status_code == 200:
    results = response.json()
    print(results)
    # for r in results.get("results", []):
    #     print(f"{r['title']}: {r['url']}")
else:
    print("Search failed:", response.status_code)
```
